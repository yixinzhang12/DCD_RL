```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load data from log0.csv
df = pd.read_csv('saved_data/log2.csv')

# Define time array assuming dt = 0.004 and 5 minutes of data
dt = 0.004  # time step in seconds
time = np.arange(0, 5 * 60, dt)[:len(df)]  # limit time to the length of the data

# Plot the data
plt.figure(figsize=(12, 6))
plt.plot(time, df['thdot'], label='theta dot')
plt.plot(time, df['th'], label='theta')
plt.plot(time, df['action'], label='Action')

# Customize the plot
plt.title("Time Series Data from log0.csv")
plt.xlabel("Time (s)")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.show()
```
